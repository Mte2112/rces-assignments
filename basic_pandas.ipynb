### 2) Re-read the data in such a way that all date columns are identified as dates and the earthquake id is used as the index

Verify that this worked using the `head` and `info` functions.


```python
df = pd.read_csv(data, parse_dates=[0, 12])
df = df.set_index('id')
```

### 4) Use `nlargest` to get the top 20 earthquakes by magnitude

Examine the structure of the `place` column. The state / country information seems to be in there. How would you get it out?

### 5) Extract the state / country using Pandas [text data functions](https://pandas.pydata.org/pandas-docs/stable/text.html)

Add it as a new column to the dataframe.

```python
state = df.place.str.split(",").str.get(1)
state
```

```python
df['state'] = state
df
```

### 6) Display each unique value from the new column

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.unique.html

```python
statevals = df.state.unique()
statevals
```

### 7) Make a bar chart of the top 5 earthquake magnitudes vs country/state 

```python
top5 = df.nlargest(5, 'mag')
top5.plot.bar(x='state', y='mag');
plt.ylabel('magnitude');
plt.title('top 5 earthquake magnitudes');
```

### 8) Create a filtered dataset that only has earthquakes of magnitude 4 or larger

```python
mag4plus = df[df.mag>=4]
mag4plus
```

### 9) Make a histogram the distribution of the Earthquake magnitudes

https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.DataFrame.hist.html

Do one subplot for the filtered and one for the unfiltered dataset.

```python
from matplotlib import pyplot as plt
%matplotlib inline
%config InLineBackend.figure_format = 'retina'
import numpy as np
```

```python
fig, ax = plt.subplots(ncols=2, nrows =1, figsize = (14,10))
df.iloc[:, 4].hist(bins=15,ax=ax[0]).set_yscale('log')
mag4plus.iloc[:, 4].hist(bins=10,ax=ax[1]).set_yscale('log')
```

### 11) Visualize the locations of earthquakes by making a scatterplot of their latitude and longitude.

Do both the filtered and unfiltered datasets. Color it by magnitude. Make it pretty

What difference do you note between the filtered and unfiltered datasets?

```python
fig, ax = plt.subplots(ncols=2, figsize=(14,5))
df.plot.scatter(x='longitude', y='latitude', c='mag', colormap='plasma', ax=ax[0])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Earthquakes")

mag4plus.plot.scatter(x='longitude', y='latitude', c='mag', colormap='plasma', ax=ax[1])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Earthquakes")
```

```python
# the data sets have different color levels. This makes sense because2 the filtered data is magnitude of >=4.0
# which means that the minimum coloring is 4.0
```
